```python
W = torch.randn(10, 1, requires_grad = True)
b = torch.randn(1, requires_grad = True)
```

```python
W
```

```python
x = torch.randn(1, 10)
```

```python
x
```

```python
output = torch.matmul(x, W) + b
```

```python
loss = 1 - output
```

```python
loss
```

```python
output
```

```python
loss.backward()
```

```python
W.grad
```

```python
b.grad
```

## Update the weights

```python
with torch.no_grad():
    W = W - 0.001 * W.grad.data
```
